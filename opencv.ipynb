```python
def show(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
cap.set(3,500)
cap.set(4,500)
while True:
    suc,img = cap.read()
    cv2.imshow("yes",img)
    if cv2.waitKey(1)&0xFF==ord('q'):
        break
```

```python
vc = cv2.VideoCapture('/home/ano/视频/Kazam_screencast_00000.mp4')
if vc.isOpened():
    ope,fra = vc.read()
else:
    ope = False
while ope:
    ret,fra = vc.read()

    if fra is None:
        break
    if ret == True:
        gray = cv2.cvtColor(fra,cv2.COLOR_BGR2GRAY)
        cv2.imshow('3',gray)
        if cv2.waitKey(40)&0xFF==27:
            break
vc.release()
cv2.destroyAllWindows()
```
