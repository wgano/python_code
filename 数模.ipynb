```python
'''
代码中的主要函数逻辑是flower，功能是绘制繁花曲线
输入的参数依次为
R：大圆半径
r:小圆半径
dis:画笔在小圆中距离边缘的距离
w:画图精度（推荐0.02以下，可缺省）
color:画笔颜色（默认红色，可缺省）
size:画笔尺寸（默认为1像素，可缺省）
新增函数逻辑rainbowflower，功能是用渐变色的画笔绘制繁花曲线
输入的参数依次为
R：大圆半径
r:小圆半径
dis:画笔在小圆中距离边缘的距离
w:画图精度（推荐0.02以下，可缺省）
times:颜色渐变的速度（默认为1，可缺省）
size:画笔尺寸（默认为1像素，可缺省）
'''

import turtle
import math
import colorsys

pie = math.pi
turtle.colormode(1.0)  #设置颜色计量方式


def rainbow_flower12(R,r,dis,w=0.01,size=1,times= 1):
    turtle.pensize(size)
    turtle.penup()
    for t in range(1001):
        dis = r*(1-k*t)
        M = R-r
        m = r-dis
        x = M*math.cos(2*pie*w*t)+m*math.cos(2*pie*w*t*R/r)
        y = M*math.sin(2*pie*w*t)+m*math.sin(2*pie*w*t*R/r)
        turtle.goto(x, y)
        color_R, color_G, color_B = colorsys.hsv_to_rgb(times*t,1.0,1.0)
        turtle.color((color_R, color_G, color_B))
        turtle.pendown()

def rainbow_flower22(R,r,dis,w=0.01,size=1,times= 1):
    turtle.pensize(size)
    turtle.penup()
    for t in range(1001):
        dis = r*(1-k*t)
        M = R+r
        m = r-dis
        x = M*math.cos(2*pie*w*t)+m*math.cos(2*pie*w*t*R/r)
        y = M*math.sin(2*pie*w*t)-m*math.sin(2*pie*w*t*R/r)
        turtle.goto(x, y)
        color_R, color_G, color_B = colorsys.hsv_to_rgb(times*t,1.0,1.0)
        turtle.color((color_R, color_G, color_B))
        turtle.pendown()
        
# flower(200,110,20,color="#154279",size=3)
# flower(200,130,110,0.01,'red',1)
# flower(140,85,5,0.01,'blue',1)

R = int(input("请输入大圆半径:"))
r = int(input("请输入小圆半径:"))
k = float(input("请输入k值:"))
lx = int(input("请输入方向和相切形式，12为内切顺时针,22为外切顺时针:"))
n = float(input("请输入n值:"))
m = float(input("请输入m值:"))
k = k*n*m
if lx == 12:
    rainbow_flower12(R,r,k,size=1,times=60)
if lx == 22:
    rainbow_flower22(R,r,k,size=1,times=60)
print("完成")

```

```python
'''
代码中的主要函数逻辑是flower，功能是绘制繁花曲线
输入的参数依次为
R：大圆半径
r:小圆半径
dis:画笔在小圆中距离边缘的距离
w:画图精度（推荐0.02以下，可缺省）
color:画笔颜色（默认红色，可缺省）
size:画笔尺寸（默认为1像素，可缺省）
新增函数逻辑rainbowflower，功能是用渐变色的画笔绘制繁花曲线
输入的参数依次为
R：大圆半径
r:小圆半径
dis:画笔在小圆中距离边缘的距离
w:画图精度（推荐0.02以下，可缺省）
times:颜色渐变的速度（默认为1，可缺省）
size:画笔尺寸（默认为1像素，可缺省）
'''

import turtle
import math
import colorsys

pie = math.pi
turtle.colormode(1.0)  #设置颜色计量方式

def lcm(a,b):
    for i in range(min(a,b),0,-1):
        if a % i == 0 and b % i == 0:
            return math.ceil(a*b/i)

def flower(R,r,dis,w=0.01,color='red',size=1):
    turtle.color(color)
    turtle.pensize(size)
    turtle.penup()
    T = math.ceil(lcm(R,r)/w/R)
    M = R-r
    m = r-dis
    for t in range(T+1):
        x = M*math.cos(2*pie*w*t)+m*math.cos(2*pie*w*t*R/r)
        y = M*math.sin(2*pie*w*t)-m*math.sin(2*pie*w*t*R/r)
        turtle.goto(x, y)
        turtle.pendown()

def rainbow_flower11(R,r,dis,w=0.01,size=1,times= 1):
    turtle.pensize(size)
    turtle.penup()
    T = math.ceil(lcm(R,r)/w/R) #向上拉入最近整数
    for t in range(T+1):
        M = R-r
        m = r-dis
        x = M*math.cos(2*pie*w*t)+m*math.cos(2*pie*w*t*R/r)
        y = M*math.sin(2*pie*w*t)-m*math.sin(2*pie*w*t*R/r)
        turtle.goto(x, y)
        color_R, color_G, color_B = colorsys.hsv_to_rgb(times*t/T,1.0,1.0)
        turtle.color((color_R, color_G, color_B))
        turtle.pendown()

def rainbow_flower12(R,r,dis,w=0.01,size=1,times= 1):
    turtle.pensize(size)
    turtle.penup()
    T = math.ceil(lcm(R,r)/w/R) #向上拉入最近整数
    for t in range(T+1):
        M = R-r
        m = r-dis
        x = M*math.cos(2*pie*w*t)+m*math.cos(2*pie*w*t*R/r)
        y = M*math.sin(2*pie*w*t)+m*math.sin(2*pie*w*t*R/r)
        turtle.goto(x, y)
        color_R, color_G, color_B = colorsys.hsv_to_rgb(times*t/T,1.0,1.0)
        turtle.color((color_R, color_G, color_B))
        turtle.pendown()
        
def rainbow_flower21(R,r,dis,w=0.01,size=1,times= 1):
    turtle.pensize(size)
    turtle.penup()
    T = math.ceil(lcm(R,r)/w/R) #向上拉入最近整数
    for t in range(T+1):
        M = R+r
        m = r-dis
        x = M*math.cos(2*pie*w*t)+m*math.cos(2*pie*w*t*R/r)
        y = M*math.sin(2*pie*w*t)+m*math.sin(2*pie*w*t*R/r)
        turtle.goto(x, y)
        color_R, color_G, color_B = colorsys.hsv_to_rgb(times*t/T,1.0,1.0)
        turtle.color((color_R, color_G, color_B))
        turtle.pendown()

def rainbow_flower22(R,r,dis,w=0.01,size=1,times= 1):
    turtle.pensize(size)
    turtle.penup()
    T = math.ceil(lcm(R,r)/w/R) #向上拉入最近整数
    for t in range(T+1):
        M = R+r
        m = r-dis
        x = M*math.cos(2*pie*w*t)+m*math.cos(2*pie*w*t*R/r)
        y = M*math.sin(2*pie*w*t)-m*math.sin(2*pie*w*t*R/r)
        turtle.goto(x, y)
        color_R, color_G, color_B = colorsys.hsv_to_rgb(times*t/T,1.0,1.0)
        turtle.color((color_R, color_G, color_B))
        turtle.pendown()
        
# flower(200,110,20,color="#154279",size=3)
# flower(200,130,110,0.01,'red',1)
# flower(140,85,5,0.01,'blue',1)

R = int(input("请输入大圆半径:"))
r = int(input("请输入小圆半径:"))
k = float(input("请输入k值:"))
dis = r*(1-k)
lx = int(input("请输入方向和相切形式，11为内切逆时针，12为内切顺时针,21为外切逆时,22为外切顺时针:"))
    
if lx == 11:
    rainbow_flower11(R,r,dis,size=1,times=60)
if lx == 12:
    rainbow_flower12(R,r,dis,size=1,times=60)
if lx == 21:
    rainbow_flower21(R,r,dis,size=1,times=60)
if lx == 22:
    rainbow_flower22(R,r,dis,size=1,times=60)
print("完成")
turtle.hideturtle()
```
