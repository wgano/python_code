```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
```

```python
x = np.array([0,20])
y = np.array([0,100])
```

```python
plt.plot(x,y)
plt.show()
```

```python
m = np.arange(-4*np.pi,4*np.pi,0.1)
n = np.sin(m)
```

```python
plt.plot(m,n,'o-.c',ms = 1,lw= 0.5)
plt.title("Test",loc = "center")
plt.grid(ls = ':',c = 'k')
plt.xlabel("x",loc = "right")
plt.ylabel("y",loc = "top")
plt.show()
```
