```python
from matplotlib import pyplot as plt
import numpy as np
import random
x = range(2,26,2)
y1 = random.sample(x, 12)
y2 = random.sample(x, 12)
```

```python
plt.plot(x,y1,'-.vg',x,y2,'-.oc')
plt.show()
```

```python
x_ = np.arange(-4*np.pi,4*np.pi,0.1) 
y_ = np.sin(x_)
plt.axhline(0,ls="-.",c="green")
plt.axvline(0,ls="-.",c="green")
plt.grid(color='r',linestyle='--',linewidth=0.5)
plt.plot(x_,y_,'-.o', ms = 5, mec = 'w')
plt.show()
```
