```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import *
%matplotlib inline
```

```python
plt.title("title")
plt.grid(axis="x",c="r")
plt.xlabel("x")
plt.ylabel("y")
plt.plot([1,2,3],[1,2,3],"o:r")
```

```python
x = np.linspace(-10,10,100)
plt.subplot(2,2,1)
plt.title("title")
plt.grid(axis="x",c="r")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.subplot(2,2,3)
plt.plot(x,x**-1)
plt.subplot(2,2,4)
plt.plot(x,x**3)
plt.show()
```

```python
a = np.array([12,54,26,42,15,63,84,49,26])
m = a**2
b = np.array(["red","green","blue","lightblue","lightgreen","pink","black","purple",0.5])
plt.scatter(a,m,s=a,c=b,alpha=0.5,cmap='virids')
plt.colorbar()
plt.show()
```

```python
x = np.array(["1", "2", "3", "4"])
y = np.array([12, 22, 6, 18])

plt.barh(x, y, color = "#4CAF50",align='edge' )
```
